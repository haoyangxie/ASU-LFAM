```python
plt.plot(mean_reward, color='blue', label='Mean Reward')
plt.fill_between(range(len(mean_reward)), min_reward, max_reward, color='blue', alpha=0.2, label='Reward Range')
plt.xlabel('Number of iterations', fontsize=20)
plt.ylabel('Average reward', fontsize=20)
plt.legend(frameon=False, fontsize=16)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.tight_layout()
# plt.savefig('../paper_plot/simplified_training_reward.png')
plt.show()
```

```python
inference_result = np.load('inference.npy', allow_pickle=True)
```

```python
random_result = np.load('random.npy', allow_pickle=True)
```

```python
x = range(1, 101)

plt.plot(x, inference_result, label=f'{round(np.mean(inference_result), 2)}')
plt.plot(x, random_result, label=f'{round(np.mean(random_result), 2)}') 


plt.legend(title='average reward')
plt.xlabel('Episode')
plt.ylabel('Reward')
plt.title('Average Reward Comparison Between Two Strategies')

plt.savefig("inference_random.png")
```

```python
current_layer = 0
current_checkpoint = 0
checkpoints_per_layer = 100
num_layers = 30
```

```python
while True:
    current_checkpoint += 1
    current_checkpoint %= checkpoints_per_layer
    print('current_checkpoint', current_checkpoint)
    if current_checkpoint == 0:
        current_layer += 1
        print('current_layer', current_layer)
    if current_layer >= num_layers:
        break
```
