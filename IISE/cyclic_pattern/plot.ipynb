```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
def extract_profile(file_path):
    df = pd.read_csv(file_path)
    df_filtered = df[[df.columns[0], df.columns[-1]]].to_numpy()
    df_filtered[:, 0] = df_filtered[:, 0] - df_filtered[0, 0]
    return df_filtered[:, 1]
```

```python
profile2 = extract_profile('2min_cyclic.csv')
profile3 = extract_profile('3min_cyclic.csv')
profile4 = extract_profile('4min_cyclic.csv')
```

```python
plt.plot(profile2[:875], linestyle='solid', label='2 minutes')
plt.plot(profile3[:878], linestyle='dashdot', label ='3 minutes')
plt.plot(profile4[:928], linestyle='dashed', label='4 minutes')
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.xlabel("Time (Seconds)", fontsize=20)
plt.ylabel("Temperature (C)", fontsize=20)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.tight_layout()
plt.savefig('../paper_plot/cycle_pattern.png')
plt.show()
```
