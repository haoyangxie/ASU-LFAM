```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_map = {}
```

```python
minutes = [1,2,3,4,6]
positions = [1,2,3,4]
path = './data/'
```

```python
df = pd.read_csv(f'{path}{1}min_position{1}.csv')
```

```python
for minute in minutes:
    for position in positions:
        df = pd.read_csv(f'{path}{minute}min_position{position}.csv')
        df_filtered = df[[df.columns[0], df.columns[-1]]].to_numpy()
        df_filtered[:, 0] = df_filtered[:, 0] - df_filtered[0, 0]
        df_map[(minute, position)] = df_filtered
```

```python
for min in minutes:
    for position in positions:
        x = df_map[(min, position)][:,0]
        y = df_map[(min, position)][:,1]
        plt.plot(x,y,label=f'{min}min_pos{position}')
    plt.legend()
    plt.show()
    plt.clf()
```
