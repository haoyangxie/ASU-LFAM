# draw for cycle pattern

```python
df = pd.read_csv('./planter_cycle.csv')
temp_list = df['Function 1 [C]'].tolist()
plt.plot(temp_list[1500:])
plt.ylim(70, 200)
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.xlabel('Time(s)', fontsize=16)
plt.ylabel('Surface temperature(°C)', fontsize=16)
new_xticks = np.linspace(0, 1400, num=8) 
new_labels = np.linspace(0, 2800, num=8) 
plt.xticks(ticks=new_xticks, labels=new_labels.astype(int), fontsize=16)
plt.yticks(fontsize=16)
plt.tight_layout()
plt.savefig('./paper_figures/cycle_pattern.png')
```

# data cleaning

```python
df = pd.read_csv('./csv/planters/transformer_input/layer_8.csv')
```

```python
list_of_lists = [df[column].tolist() for column in df]
```

```python
data = np.array(list_of_lists)
```

```python
location = np.linspace(0, 743, 743)

# Assuming time ranges from 0 to some max time and we have 120 time points
time = np.linspace(0, 120, 120)

# Create meshgrid for plotting
time_grid, location_grid = np.meshgrid(time, location)

# Set up the figure and 3D axis
fig = plt.figure(figsize=(14, 6))
ax = fig.add_subplot(111, projection='3d')

# Create a surface plot using the meshgrid and the temperature data
surf = ax.plot_surface(time_grid, location_grid, data, cmap='jet')

# Add color bar which maps values to colors
cbar = fig.colorbar(surf, shrink=0.5, aspect=5)
cbar.set_label('Temperature (°C)')

# Set axis labels
ax.set_xlabel('Time (s)')
ax.set_ylabel('Location')
ax.set_zlabel('Temperature (°C)')

# Show plot
plt.show()
```
